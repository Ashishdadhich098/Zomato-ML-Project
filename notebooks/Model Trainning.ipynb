```python
import geopy 
from geopy.distance import geodesic


# calculate distance

def calculate_distance(raw):
    restaurant_cords=(raw['Restaurant_latitude'],raw['Restaurant_longitude'])
    delivery_cords=(raw['Delivery_location_latitude'],raw['Delivery_location_longitude'])
    return geodesic(restaurant_cords,delivery_cords).kilometers

df['distance (in km)']=df.apply(calculate_distance , axis=1)

df.head()
```

```python
# Define which columns should be ordinal-encoded(categorical) and which should be scaled(numerical)
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(exclude='object').columns
```

```python
print(categorical_cols)
print (numerical_cols)
```

```python
# Define the custom ranking for each ordinal variable
Weather_conditions_categories=['Sunny','Stormy','Sandstorms','Windy','Fog','Cloudy']
Road_traffic_density_categories=['Low','Medium','High','Jam']
Type_of_vehicle_categories=['scooter','electric_scooter','bicycle','motorcycle']
Festival_categories=['No','Yes']
City_categories=['Urban','Metropolitian','Semi-Urban']
```

```python
from sklearn.impute import SimpleImputer ## Handling Missing Values
from sklearn.preprocessing import StandardScaler # Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
## Numerical Pipeline
num_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())

    ]

)

# Categorigal Pipeline
cat_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('ordinalencoder',OrdinalEncoder(categories=[Weather_conditions_categories,Road_traffic_density_categories,Type_of_vehicle_categories,Festival_categories,City_categories])),
    ('scaler',StandardScaler())
    ]

)

preprocessor=ColumnTransformer([
('num_pipeline',num_pipeline,numerical_cols),
('cat_pipeline',cat_pipeline,categorical_cols)
])
```

```python
# Train test split 

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.40,random_state=40)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
import numpy as np

# Define the evaluate_model function if not already defined
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square

# Train multiple models
models = {
    'LinearRegression': LinearRegression(),
    'Lasso': Lasso(),
    'Ridge': Ridge(),
    'Elasticnet': ElasticNet(),
    'RandomForestRegressor': RandomForestRegressor(),
    'SVR': SVR(),
    'XGBRegressor': XGBRegressor()
}

trained_model_list = []
model_list = []
r2_list = []

for model_name, model in models.items():
    model.fit(X_train, y_train)

    # Make Predictions
    y_pred = model.predict(X_test)

    # Use the evaluate_model function
    mae, rmse, r2_square = evaluate_model(y_test, y_pred)

    print(model_name)
    model_list.append(model_name)

    print('Model Training Performance')
    print("RMSE:", rmse)
    print("MAE:", mae)
    print("R2 score", r2_square * 100)

    r2_list.append(r2_square)

    print('=' * 35)
    print('\n')
```

```python
def print_best_model_and_r2(model_list, r2_list):
    best_model_name = model_list[np.argmax(r2_list)]
    best_r2_score = 100 * max(r2_list)

    print(f"Best Model: {best_model_name}")
    print(f"Best R2 Score: {best_r2_score:.2f}")

# Example usage
print_best_model_and_r2(model_list, r2_list)
```
